```python
import numpy as np
import pandas as pd
import itertools
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer,HashingVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("news.csv", index_col=None)
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
y = df['label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(df['text'],y,test_size=0.33,random_state=53)
```

### Using CountVector to convert the categorial to numerical value
- finding the unique values first
- then counting the unique words in each line or row
- then assigning the values 

```python
count_vectorizer = CountVectorizer(stop_words='english')
count_train = count_vectorizer.fit_transform(x_train)
print(count_train)
count_test = count_vectorizer.transform(x_test)
```

```python
# Finding the featurerized Unique values in the text column
total_uniqueValues = len(count_vectorizer.get_feature_names_out())
```

```python
total_uniqueValues
```

```python
# Model creation: 
clf = MultinomialNB()

clf.fit(count_train, y_train)
pred = clf.predict(count_test)
```

```python
# finding the metrics score:
score = metrics.accuracy_score(y_test,pred)
print("accuracy: %0.3f"%score)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test,pred)
```

```python
print(report)
```

### Save the Model and Vectorizer

```python
import joblib

# Save the vectorizer
joblib.dump(count_vectorizer, 'count_vectorizer.pkl')

# Save the trained model
joblib.dump(clf, 'fake_news_model.pkl')
```

### Load the Saved Model and Vectorizer

## 2nd Method for text preprocessing
## **TF-IDF Vector**

```python
tfidf_vector = TfidfVectorizer(stop_words='english')

tf_count_train = tfidf_vector.fit_transform(x_train)
print(tf_count_train)
tf_count_test = tfidf_vector.transform(x_test)
```

```python
# Model creation: 
tf_clf = MultinomialNB()

tf_clf.fit(tf_count_train, y_train)
tf_pred = tf_clf.predict(tf_count_test)
```

```python
# finding the metrics score:
tf_score = metrics.accuracy_score(y_test,tf_pred)
print("accuracy: %0.3f"%tf_score)
```
