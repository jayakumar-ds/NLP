#### polarity_scores() is a function from the VADER sentiment analysis tool in the nltk library.

üîπ VADER = Valence Aware Dictionary and sEntiment Reasoner
üîπ It's made for short texts,
- Use **compound** to decide the final sentiment:

- compound >= 0.05 ‚Üí Positive

- compound <= -0.05 ‚Üí Negative

- Otherwise ‚Üí Neutral

### Creating python function for above examples:

```python
def sentimentAnalysis(data, columnname):
    import matplotlib.pyplot as plt
    import seaborn as sns
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    import pandas as pd

    # Initialize sentiment analyzer
    sid = SentimentIntensityAnalyzer()

    # Drop rows with null values in the target column
    data = data.dropna(subset=[columnname])

    # Apply sentiment analysis
    data['scores'] = data[columnname].apply(lambda text: sid.polarity_scores(str(text)))

    # Extract compound score
    data['compound'] = data['scores'].apply(lambda x: x['compound'])

    # Categorize sentiment based on compound score
    data['sentiment'] = data['compound'].apply(
        lambda score: 'positive' if score >= 0.05 else ('negative' if score <= -0.05 else 'neutral')
    )

    # Plot sentiment distribution
    plt.figure(figsize=(6, 4))
    sns.countplot(data=data, x='sentiment', hue='sentiment', palette='viridis', legend=False)
    plt.title("Sentiment Distribution")
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    plt.show()

    return data
```

```python
sentimentAnalysis(dataset,'Chat')
```

## **Using custom stopwords to add addition stopword on my own**

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

default_stopwords = set(stopwords.words('english'))
custom_stopwords = default_stopwords.union({"please", "hi", "ok", "thanks",","})

text = "hi there, I need help please with my data"
words = nltk.word_tokenize(text)

filtered_words = [word for word in words if word.lower() not in custom_stopwords]

print(filtered_words)
```
